```python
df.drop(['Country Code','Indicator Name'],axis=1,inplace=True)
```

```python
Extraction of top-10 countries with respect to total population
```

```python
# Filter data for total population
total_population_data = df[df['Indicator Code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2023", ascending=False)

# Get the top ten countries with the highest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Name']] )
```

```python
import matplotlib.pyplot as plt

# List of countries to plot
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Define non-country names explicitly
non_countries = ['World', 'IDA & IBRD total', 'Low & middle income', 'Middle income', 
                 'IBRD only', 'Early-demographic dividend', 'Lower middle income', 
                 'Upper middle income', 'East Asia & Pacific', 'Late-demographic dividend']

# Filter data for countries
filtered_countries = df[df['Country Name'].isin(countries_to_plot)]

# Filter data for non-countries
filtered_non_countries = df[df['Country Name'].isin(non_countries)]

# Plot for countries
plt.figure(figsize=(10, 5))
plt.bar(filtered_countries['Country Name'], filtered_countries['2022'])
plt.xlabel('Country Name')
plt.ylabel('Population in 2022')
plt.title('Population of Selected Countries in 2022')
plt.xticks(rotation=45, ha='right')  # Rotating the x-axis labels for readability
plt.show()

# Plot for non-countries
plt.figure(figsize=(10, 5))
plt.bar(filtered_non_countries['Country Name'], filtered_non_countries['2022'])
plt.xlabel('Non-Country Name')
plt.ylabel('Population in 2022')
plt.title('Population of Non-Country Groups in 2022')
plt.xticks(rotation=45, ha='right')  # Rotating the x-axis labels for readability
plt.show()
```
